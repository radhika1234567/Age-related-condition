```python
import seaborn as sns
```

```python
plt.figure(figsize = (12, 17), dpi = 100)
corr_train = df.drop(columns = ['EJ'], axis = 1).corr()
heatmap = sns.heatmap (corr_train[['Class']].sort_values (by = 'Class', ascending = False),
                             vmin = -1, vmax = 1, annot = True)
heatmap.set_title ('Features Correlating with Class', fontdict = {'fontsize':20}, pad = 18,fontweight = 'bold');
```

```python
import plotly.express as px
```

```python
df.drop(columns=["Id"],inplace=True)
```

```python
def create_boxplot(data,x,y):
    fig = px.box(data, x=x, y=y, color = x, title = f"Box Plots\n{x} vs {y}")
    fig.show()

for cols in df.columns:
    if cols!="Class" and cols!="EJ":
        create_boxplot(data=df,y=cols,x="Class")
```
